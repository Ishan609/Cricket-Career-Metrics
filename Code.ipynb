```python
def rohit_ODI():
        
        win=Tk()
        win.title("ROHIT SHARMA_ODI")
        win.state('zoomed')
        win.configure(bg="blue")
        lbl_title=Label(win,text="Rohit Sharma ODI Insights",font=('',55,'bold','underline'),bg='orange')
        lbl_title.pack()
        
        df=pd.read_excel("E://sachin/rohit/t20.xlsx")
        df10=pd.read_excel("E://sachin/rohit/odi.xlsx")
        df["Dismissal"]=df10.Dismissal
        df['Year']=df['Start Date'].dt.year

        df.drop(["Bat1","Bat2"],axis=1)
        df.drop(["Bat1","Bat2"],axis=1)
        df.drop(["Bat1","Bat2","Column1"],axis=1)
        df.Runs.replace("[^0-9]","DNB",regex=True,inplace=True)

        df[['Match Type', 'Opposition']] = df['Opposition'].str.split(' v ', expand=True)
        df.drop(["Bat1","Bat2","Column1"],axis=1)
        df1=df[df["Match Type"]=="ODI"]
        df2=df1[(df1.Runs!="DNB")&(df1.Runs!="TDNB")]
        df2.isnull().sum()
        df2.Runs=pd.to_numeric(df2.Runs)
        df2["4s"]=pd.to_numeric(df2["4s"])
        df2["6s"]=pd.to_numeric(df2["6s"])
        df2.SR=pd.to_numeric(df2.SR)
        a=df2.Runs.sum()
        label1=Label(win,text=f"1. Total runs scored by Rohit Sharma in ODI {a}",bg="blue",fg="White",font=("Time New Roman",13,"bold")).place(x=100,y=130)
        b=df2["4s"].sum()
        label2=Label(win,text=f"2. Total numbers of 4 Rohit Sharma in ODI {b}",bg="blue",fg="White",font=("Time New Roman",13,"bold")).place(x=100,y=150)
        c=df2["6s"].sum()
        label3=Label(win,text=f"3. Total numbers of 6 Rohit Sharma hitted in ODI {c}",bg="blue",fg="White",font=("Time New Roman",13,"bold")).place(x=100,y=170)

        duck=df2[df2.Runs==0].shape
        label4=Label(win,text=f"4. Number of times Rohit Sharma got out on duck is {duck[0]}",bg="blue",fg="White",font=("Time New Roman",13,"bold")).place(x=100,y=190)
        filtered_df2=df2[df2.Runs==0]
        country_duck=filtered_df2.Opposition.value_counts().reset_index().values
        label5=Label(win,text=f"5. Maximum times a country Rohit Sharma got out on duck in his career is {country_duck[0][0]}  i.e {country_duck[0][1]} ",bg="blue",fg="White",font=("Time New Roman",13,"bold")).place(x=100,y=210)
        label6=Label(win,text=f"6. Average strike rate in ODI is {df2.SR.mean()}",bg="blue",fg="White",font=("Time New Roman",13,"bold")).place(x=100,y=230)
        label7=Label(win,text=f"7. Maximum strike rate of Rohit Sharma in his career is {df2.SR.max()}",bg="blue",fg="White",font=("Time New Roman",13,"bold")).place(x=100,y=250)
        max_score=df2.Runs.max()
        max_scor_df2=df2[df2.Runs==max_score]
        max_scor=max_scor_df2.Opposition.reset_index().values
        label8=Label(win,text=f"8. Maximum run scored by Rohit Sharma in his career in 1 innings is {max_score} against {max_scor[0][1]}",bg="blue",fg="White",font=("Time New Roman",13,"bold")).place(x=100,y=270)
        label9=Label(win,text=f"9. Number of centuries made by Rohit Sharma in his ODI career is {df2[df2.Runs>=100].shape[0]}",bg="blue",fg="White",font=("Time New Roman",13,"bold")).place(x=100,y=290)
        centuries_df2=df2[df2.Runs>=100]
        no_centuries=centuries_df2.Opposition.value_counts().reset_index().values
        label10=Label(win,text=f"10. Maximum number of centuries made by Rohit Sharma in his ODI career is {no_centuries[0][1]} against {no_centuries[0][0]} ",bg="blue",fg="White",font=("Time New Roman",13,"bold")).place(x=100,y=310)
        label11=Label(win,text=f"11. Number of fifties Rohit Sharma Scored in his ODI career is {df2[(df2.Runs>=50)].shape[0]}",bg="blue",fg="White",font=("Time New Roman",13,"bold")).place(x=100,y=330)
        fifties_df2=df2[(df2.Runs>=50)&(df2.Runs<=99)]
        no_fifties=fifties_df2.Opposition.value_counts().reset_index().values
        label12=Label(win,text=f"12. Maximum number of fifties Rohit Sharma made against a country in his career is {no_fifties[0][1]} against {no_fifties[0][0]} ",bg="blue",fg="White",font=("Time New Roman",13,"bold")).place(x=100,y=350)
        conversion_df2=df2[(df2.Runs>=50)&(df2.Runs<=99)]
        label13=Label(win,text=f"13. Number of times Rohit Sharma fails to convert his fifties into century {conversion_df2.shape[0]}",bg="blue",fg="White",font=("Time New Roman",13,"bold")).place(x=100,y=370)
        conversion=conversion_df2.Opposition.value_counts().reset_index().values
        label14=Label(win,text=f"14. Maximum number of times Rohit Sharma convert his fifty into century in his ODI career is {conversion[0][1]} against {conversion[0][0]}",bg="blue",fg="White",font=("Time New Roman",13,"bold")).place(x=100,y=390)
        image_path = "E:/sachin/ROHIT1.jpg"  
        img=Image.open(image_path)
        imgtk=ImageTk.PhotoImage(img,master=win)
        lb=Label(win,image=imgtk)
        lb.place(x=1020,y=110)
        dfpvt1=df2.pivot_table(index="Year",values="Runs",aggfunc="sum")
        dfpvt1.plot(kind='line',figsize=(15,6),legend=False,marker='o')
        plt.grid()
        plt.xlabel('Years')
        plt.ylabel('Runs')
        plt.title('Total Runs Scored In Each Year')
        plt.xticks(df.Year.unique())
        plt.savefig('E:/sachin/rohit_odi_line_chart.png')
        dfpvt2=df2.pivot_table(index="Opposition",values="Runs",aggfunc="sum")
        dfpvt2.plot(kind='bar', color='blue',legend=False,figsize=(15,6))
        plt.xlabel('Opposition')
        plt.ylabel('Total Runs')
        plt.title('Total Runs Scored Against Each Opposition')
        plt.xticks(rotation=45, ha='right')
        plt.savefig('E:/sachin/rohit_odi_bar_chart.png')
        image_path_pvt1 = 'E:/sachin/rohit_odi_line_chart.png'
        resize_img(image_path_pvt1)
        img_pvt1=Image.open(image_path_pvt1)
        imgtk_pvt1=ImageTk.PhotoImage(img_pvt1,master=win)
        lb_pvt1=Label(win,image=imgtk_pvt1)
        lb_pvt1.place(x=50,y=430)
        image_path_pvt2 = 'E:/sachin/rohit_odi_bar_chart.png'
        resize_img(image_path_pvt2)
        img_pvt2=Image.open(image_path_pvt2)
        imgtk_pvt2=ImageTk.PhotoImage(img_pvt2,master=win)
        lb_pvt2=Label(win,image=imgtk_pvt2)
        lb_pvt2.place(x=780,y=430) 
        button=Button(win,text="HOME",font=("Time New Roman",20,"bold"),command=main_win).place(x=1380,y=5,height=60,width=150)
        win.mainloop()
```
